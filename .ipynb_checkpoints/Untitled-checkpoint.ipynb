```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import confusion_matrix
from xgboost.sklearn import XGBClassifier   
%matplotlib inline
```

```python
simuTrain = pd.read_pickle('data/dataSimu')
X_train, X_test, y_train, y_test= train_test_split(simuTrain.drop('label', axis=1), simuTrain.label, test_size=0.5)
```

```python
xgbc = XGBClassifier(              
                max_depth= 1,      
                learning_rate= 0.5,
                n_estimators= 100, 
                booster= 'gbtree', 
                gamma= 0.5,        
                n_jobs= 15,        
                reg_lambda= 0.1,   
                subsample= 0.20,   
                eval_metric= 'auc' 
                )  
xgbc.fit(X_train, y_train)
```

```python
TN = []
FP = []
FN = []
TP = []
for x in np.linspace(0,1,20):
    y_pred = xgbc.predict_proba(X_test)[:,1]
    cut = x
    y_pred[y_pred>=cut]=1
    y_pred[y_pred<cut]=0
    [[tn,fp],[fn,tp]] = confusion_matrix(y_test,y_pred)
    TN= np.append(TN, tn)
    FP= np.append(FP, fp)
    FN= np.append(FN, fn)
    TP= np.append(TP, tp)
#print('tp: %s, fn: %s, fp: %s, tn: %s ' %(tp, fn, fp, tn))
#print('N_on: %s, N_off: %s' %(tp+fp, fn+tn))
```

```python
N_off= TN + FP
N_on= TP+ FN+ FP+ TN
plt.plot(np.linspace(0,1,20), TN, label='tn')
plt.plot(np.linspace(0,1,20), FN, label='fn')
plt.plot(np.linspace(0,1,20), TP, label='tp')
plt.plot(np.linspace(0,1,20), FP, label='fp')
plt.legend()
```

```python
plt.plot(np.linspace(0,1,20), FP+TP, label='fp')
plt.legend()
```

```python
N_off= TN-FN
N_on= TP- FP+ TN-FN
plt.plot(np.linspace(0,1,20), N_off, label='On')
plt.plot(np.linspace(0,1,20), N_on, label='Off')
plt.legend()
```

```python
plt.plot(np.linspace(0,1,20),N_on/np.sqrt(N_on+N_off)-N_off/np.sqrt(N_on+N_off))
```

```python
recall = []
hits = []
precission = []
for x in np.linspace(0,1,40):
    y_pred_xgbc = xgbc.predict_proba(X_test)[:,1]
    cut = x
    y_pred_xgbc[y_pred_xgbc>=cut]=1
    y_pred_xgbc[y_pred_xgbc<cut]=0
    ma = confusion_matrix(y_test, y_pred_xgbc)
    hits.append(np.sum(y_pred_xgbc>=cut))
    recall.append(ma[1,1]/(ma[1,1]+ma[1,0]))
    precission.append(ma[1,1]/(ma[1,1]+ma[0,1]))
```

```python
plt.plot(np.linspace(0,1,40), recall[:40], label= 'recall= TP/(TP+FN)')
plt.plot(np.linspace(0,1,40), precission[:40], label= 'precission = TP/(TP+FP)')
plt.plot(np.linspace(0,1,40), np.sqrt(hits[:40]/max(hits)), label= 'Testset size')
plt.legend()
```

```python
plt.plot(np.linspace(0,1,40), precission*np.sqrt(hits[:40]), label= 'Testset size')
```
