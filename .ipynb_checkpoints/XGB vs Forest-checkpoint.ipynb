```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import roc_auc_score
from fact.io import read_h5py
%matplotlib inline
plt.style.use('ggplot')
```

```python
features = ['conc_core', 'concentration_one_pixel', 'concentration_two_pixel',
            'leakage', 'leakage2', 'size', 'width', 'num_islands',
            'num_pixel_in_shower', 'ph_charge_shower_max', 'ph_charge_shower_mean',
            'ph_charge_shower_min', 'ph_charge_shower_variance', 'theta_deg']
off_data = ['theta_deg_off_1', 'theta_deg_off_2', 'theta_deg_off_3', 'theta_deg_off_4',
            'theta_deg_off_5']
```

```python
crab_data = read_h5py(
    '/home/msackel/Desktop/gammaClassification/data/raw_data/crab_precuts.hdf5',
    key='events',
    columns=list(features)+list(off_data)
)
```

```python
crabTrain = pd.read_pickle('data/crabTrain')
```

```python
from xgboost.sklearn import XGBClassifier
reguEsti = {
    'max_depth':4,
    'learning_rate':0.3,
    'colsample_bytree':0.5, 
    'subsample': 0.3,
    'silent':0, 
    'objective':'binary:logistic',         
    'eval_metric':'merror',
    'nthread':20, 
    'num_round':200
}
bestEsti = {
    'max_depth':5,
    'learning_rate':1,
    'gamma':10,
    'booster':'gbtree',
    'silent':0, 
    'nthread':20 ,
    'subsample': 0.3,
    'objective':'binary:logistic',         
    'eval_metric':'auc',
}
xgbc = XGBClassifier(**bestEsti)
xgbc.fit(crabTrain.drop('label', axis=1), crabTrain.label)
```

```python
Tree = RandomForestClassifier(n_estimators=100, n_jobs=15, max_depth=15, max_features=6, bootstrap=True)
Tree.fit(crabTrain.drop('label', axis=1), crabTrain.label)
```

```python
print(Tree.score(crabTrain.drop('label', axis=1), crabTrain.label))
print(xgbc.score(crabTrain.drop('label', axis=1), crabTrain.label))
```

```python
crab_data['pred_xgbc'] = xgbc.predict_proba(crab_data[features].drop('theta_deg', axis=1))[:,1]
crab_data['pred_tree'] = Tree.predict_proba(crab_data[features].drop('theta_deg', axis=1))[:,1]
```

```python
theta_off = pd.concat([crab_data['theta_deg_off_' + str(i)] for i in range(1, 6)])
```

# Threshold of both estimators measured on labeled data

```python
from sklearn.metrics import recall_score, precision_score

crab_train = pd.read_pickle('data/train_test_split/CrabTrain')
crab_test = pd.read_pickle('data/train_test_split/CrabTest')

Tree.fit(crab_train.drop('label', axis=1), crab_train.label)
xgbc.fit(crab_train.drop('label', axis=1), crab_train.label)
thresholds = np.linspace(0, 1, 50)
precisions = []
Precisions = []
recalls = []
Recalls = []

for threshold in thresholds:
    pred_tree = Tree.predict_proba(crab_test.drop('label',axis=1))[:,1]
    pred_xgbc = xgbc.predict_proba(crab_test.drop('label',axis=1))[:,1]
    pred_tree[pred_tree<= threshold] = 0
    pred_tree[pred_tree>= threshold] = 1
    pred_xgbc[pred_xgbc<= threshold] = 0
    pred_xgbc[pred_xgbc>= threshold] = 1
    precisions.append(precision_score(crab_test.label.values, pred_tree))
    Precisions.append(precision_score(crab_test.label.values, pred_xgbc))
    recalls.append(recall_score(crab_test.label.values, pred_tree))
    Recalls.append(recall_score(crab_test.label.values, pred_xgbc))
```

```python
plt.plot(thresholds, precisions, '-.', label='Tree precision')
plt.plot(thresholds, recalls, '--', label='Tree recall')
plt.plot(thresholds, Precisions, label='XGBC precision')
plt.plot(thresholds, Recalls, label='XGBC recall')
plt.xlabel('threshold')
plt.legend(loc='best')
```
